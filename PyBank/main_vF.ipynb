```python
# We have to open the file
from pathlib import Path
import csv
csvpath = Path("budget_data.csv")

# Initialize list of records
records = []
changes_list = []
month_list = []

# read the file
with open(csvpath) as data_1:
    read = csv.reader(data_1, delimiter = ',')
    #print(read)
    header = next(read)
    #print(header)
    # Append the column 'Average' to the header
    header.append("Average")
    # Append the header to the list of records
    records.append(header)
        
    num_rows = 1
    row_1 = next(read)
    records.append(row_1)
    #print(row_1)
    totals = int(row_1[1])
    #print(int(row_1[1]))
    previous_month = int(row_1[1])
    #print(previous_month)
    average = 0
   
    for rows in read:
        num_rows += 1
        totals += int(rows[1])
        
        #net change current month minus previous month and then add it to the change_list
        net_change = int(rows[1]) - previous_month
        previous_month = int(rows[1])
        changes_list.append(net_change)
        month_list.append(rows[0])
        
        #print(month_list)
        # Append the average to the row
        rows.append(net_change)
        # Append the row to the list of records
        records.append(rows)
        
    # Calculate the average and round to the nearest 2 decimal places
    average = round(sum(changes_list) / (num_rows - 1), 2)

    High = max(changes_list)
    Low = min(changes_list)
    
    High_Index = changes_list.index(High)
    Low_Index = changes_list.index(Low)

    High_M = month_list[High_Index]
    Low_M = month_list[Low_Index]

# Set the path for the output.csv
output = Path("output.csv")

# Open the output path as a file and pass into the 'csv.writer()' function
# Set the delimiter/separater as a ','
with open(output, 'w') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter = ',')

    # Loop through the list of records and write every record to the
    # output csv file
    for record in records:
        csvwriter.writerow(record)
        
# Set the path for the output.csv
#output = Path("output.csv")

print("********Financial Analysis********")
print("----------------------------------------------")
#The total number of months included in the dataset.
print("Total Months: " + str(num_rows))

#The net total amount of Profit/Losses over the entire period.
print("Total: $" + str(totals))
    
#The average of the changes in Profit/Losses over the entire period.
print("Average Change: $" + str (round (average, 2)))

#The greatest increase in profits (date and amount) over the entire period.
print("Greatest Increase in Profits: " + High_M + " ($"+ str(High) +")")


#The greatest decrease in losses (date and amount) over the entire period.
print("Greatest decrease in Losses: " + Low_M + " ($"+ str(Low) +")")

#Your resulting analysis should look similar to the following:
#print(f"Greatest decrease in Losses: {Low_M} (${str(High)})")

#Total Months: 86
#Total: $38382578
#Average  Change: $-2315.12
#Greatest Increase in Profits: Feb-2012 ($1926159)
#Greatest Decrease in Profits: Sep-2013 ($-2196167)
```
